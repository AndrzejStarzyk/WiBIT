```python
data = pd.read_csv('resources/dates.csv', sep=';')
data
```

```python
data.iloc[0][0]
```

```python
import spacy
```

```python
nlp = spacy.load("pl_core_news_lg")

```

```python
for i in range(len(data)):
    doc = nlp(data.iloc[i][0])
    print("--- "+doc.text)
    for tok in doc:
        print(tok.text, tok.pos_, tok.ent_type_)
    for ent in doc.ents:
        print(ent.text, ent.start_char, ent.end_char, ent.label_, ent.start)
```

```python
patterns = [[
        {"ENT_TYPE": "date"},
        {"LOWER": "-", "OP": "?"},
        {"POS": 'ADJ', "OP": "?"},
        {"POS": "NOUN", "ENT_TYPE":"date"}
    ],
    [
        {"POS": 'ADJ', "ENT_TYPE": "date"},
        {"LOWER": "do", "OP": "?"},
        {"POS": 'ADJ', "OP": "?"},
        {"POS": "NOUN", "ENT_TYPE":"date"}
    ]
]
```

```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)
matcher.add("DatePhrase", patterns)
for i in range(len(data)):

    doc = nlp(data.iloc[i][0])
    print("--- "+doc.text)
    matches = matcher(doc)
    for m_id, start, end in matches:
        print(doc[start:end])
```

```python
import pandas as pd
data = pd.read_csv('resources/answers.csv')
data.iloc[0][1]
```

```python
from transformers import GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
enc = tokenizer(data.iloc[0][1])

for i in enc['input_ids']:
    print(tokenizer.decode(i))
```

```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
enc = tokenizer(data.iloc[0][1])
for i in enc['input_ids']:
    print(tokenizer.decode(i))
```
